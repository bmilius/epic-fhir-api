```python
groups = {"inpatient_all"}
for index, item in enumerate(codechunks):
    codestr = ",".join(item)
    for group in groups:
        print(group, end='')
        patient_file = "patients_"+group+".txt"
        outfileprefix = "Observations/"+group+"_"+resource+"_"+str(index)+"_"
        with open(patient_file,'r') as infile:
            patients = infile.readlines()
        for patient in patients:
            patient = patient.strip()
    #         print(patient)
            url = (baseurl + "&patient=" + patient + "&category=laboratory" 
                + "&code=http://loinc.org|" + codestr.strip()
                  )
            response = requests.get(url=url, headers=headers)
            print(response.status_code)
            if response.status_code == 200:
                if json.loads(response.text)['total'] != 0:
                    fname = outfileprefix+patient+'.json'
                    with open(fname, 'w') as outfile:
                        json.dump(response.json(), outfile, indent=4)
                    print(".", end='')
        print("")
print("done")
#         print("patient json written to file {fn}".format(fn=fname))
```
